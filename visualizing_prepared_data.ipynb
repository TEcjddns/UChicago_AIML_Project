```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/after prep.csv')

# def count_plot(field, df):
#     sns.countplot(df, x=field)
#     plt.title(field)
```

```python
categorical_columns = ['Overall_rating', 'work_life_balance', 'skill_development', 'salary_and_benefits', 'job_security', 'career_growth', 'work_satisfaction']

fig, ax =plt.subplots(1, 7)
fig.set_size_inches(20, 5)
for index, column in enumerate(categorical_columns):
    sns.countplot(df, x=column, ax=ax[index])
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# stacked bar chart of each rating group 1,2,3,4,5
groups = ['Overall_rating', 'work_life_balance', 'skill_development', 'salary_and_benefits']


values1  = []
values2  = []
values3  = []
values4  = []
values5  = []

# for loop over groups
for group in groups:
    values = df[group].groupby(df[group]).count()
    values1.append(values.values[0])
    values2.append(values.values[1])
    values3.append(values.values[2])
    values4.append(values.values[3])
    values5.append(values.values[4])


fig, ax = plt.subplots()

# Stacked bar chart
ax.bar(groups, values1)
ax.bar(groups, values2, bottom = values1)
ax.bar(groups, values3, bottom = values2)
ax.bar(groups, values4, bottom = values3)
ax.bar(groups, values5, bottom = values4)
```
